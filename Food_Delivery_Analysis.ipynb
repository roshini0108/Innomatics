```python
import pandas as pd

# load final dataset
df = pd.read_csv("final_food_delivery_dataset.csv")


df["order_date"] = pd.to_datetime(df["order_date"])


gold_city = (
    df[df["membership"]=="Gold"]
    .groupby("city")["total_amount"]
    .sum()
    .idxmax()
)
print("1Ô∏è‚É£ Highest Gold revenue city:", gold_city)


cuisine_avg = df.groupby("cuisine")["total_amount"].mean().idxmax()
print("2Ô∏è‚É£ Highest avg order value cuisine:", cuisine_avg)



user_totals = df.groupby("user_id")["total_amount"].sum()
count_users = (user_totals > 1000).sum()
print("3Ô∏è‚É£ Users spending > ‚Çπ1000:", count_users)


bins = [3.0,3.5,4.0,4.5,5.0]
labels = ["3.0‚Äì3.5","3.6‚Äì4.0","4.1‚Äì4.5","4.6‚Äì5.0"]

df["rating_range"] = pd.cut(df["rating"], bins=bins, labels=labels)

best_rating = df.groupby("rating_range")["total_amount"].sum().idxmax()
print("4Ô∏è‚É£ Highest revenue rating range:", best_rating)


gold_city_avg = (
    df[df["membership"]=="Gold"]
    .groupby("city")["total_amount"]
    .mean()
    .idxmax()
)
print("5Ô∏è‚É£ Gold highest avg order value city:", gold_city_avg)


rest_count = df.groupby("cuisine")["restaurant_id"].nunique()
revenue = df.groupby("cuisine")["total_amount"].sum()

combo = pd.DataFrame({"restaurants":rest_count, "revenue":revenue})
fewest = combo.sort_values(["restaurants","revenue"], ascending=[True,False]).index[0]
print("6Ô∏è‚É£ Fewest restaurants but good revenue cuisine:", fewest)


pct = (len(df[df["membership"]=="Gold"]) / len(df)) * 100
print("7Ô∏è‚É£ % orders by Gold members:", round(pct), "%")


name_col = [c for c in df.columns if "restaurant_name" in c][0]

rest_stats = df.groupby(name_col).agg(
    avg=("total_amount","mean"),
    count=("order_id","count")
)

answer8 = rest_stats[rest_stats["count"] < 20]["avg"].idxmax()

print("8Ô∏è‚É£ Restaurant:", answer8)

combo_rev = (
    df.groupby(["membership","cuisine"])["total_amount"]
    .sum()
    .idxmax()
)
print("9Ô∏è‚É£ Highest revenue combination:", combo_rev)



df["quarter"] = df["order_date"].dt.quarter
best_q = df.groupby("quarter")["total_amount"].sum().idxmax()
print("üîü Highest revenue quarter: Q", best_q)


```

```python
import pandas as pd

df = pd.read_csv("final_food_delivery_dataset.csv")


df["order_date"] = pd.to_datetime(df["order_date"], dayfirst=True)


gold_orders = len(df[df["membership"]=="Gold"])
print("1Ô∏è‚É£ Gold total orders:", gold_orders)


hyd_revenue = round(
    df[df["city"]=="Hyderabad"]["total_amount"].sum()
)
print("2Ô∏è‚É£ Hyderabad total revenue:", hyd_revenue)


distinct_users = df["user_id"].nunique()
print("3Ô∏è‚É£ Distinct users:", distinct_users)

gold_avg = round(
    df[df["membership"]=="Gold"]["total_amount"].mean(), 2
)
print("4Ô∏è‚É£ Gold average order value:", gold_avg)


rating_orders = len(df[df["rating"] >= 4.5])
print("5Ô∏è‚É£ Orders with rating ‚â• 4.5:", rating_orders)


top_city = (
    df[df["membership"]=="Gold"]
    .groupby("city")["total_amount"]
    .sum()
    .idxmax()
)

orders_top_city = len(
    df[(df["membership"]=="Gold") & (df["city"]==top_city)]
)

print("6Ô∏è‚É£ Orders in top Gold revenue city:", orders_top_city)
print("   (City:", top_city, ")")

```
